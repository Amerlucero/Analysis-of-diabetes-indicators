1.Selection and review of the database

```python
df = pd.read_csv("diabetes_binary_5050split_health_indicators_BRFSS2015.csv")

df
```

2.Selection of the columns 

```python
columnas_seleccionadas= df[["Diabetes_binary", "HighBP","HighChol","CholCheck",	"BMI","Smoker","PhysActivity"	,"Sex","Age"]]
columnas_seleccionadas
```

3.Data count of the number of patients with Diabetes and which do not according to their gender

```python
count_sex = df.groupby(['Diabetes_binary','Sex']).size().reset_index(name = 'Count')
count_sex
```

4.Creation of 4 groups, women and men who suffer, women and men who do not suffer.

```python

diabetes_mujeres = columnas_seleccionadas[(columnas_seleccionadas["Diabetes_binary"] == 1) & (columnas_seleccionadas['Sex'] == 0)]
diabetes_mujeres

```

```python
diabetes_hombres = columnas_seleccionadas[(columnas_seleccionadas["Diabetes_binary"] == 1) & (columnas_seleccionadas['Sex'] == 1)]
diabetes_hombres

```

```python
no_diabetes_mujeres = columnas_seleccionadas[(columnas_seleccionadas["Diabetes_binary"] == 0) & (columnas_seleccionadas['Sex'] == 0)]
no_diabetes_mujeres
```

```python
no_diabetes_hombres = columnas_seleccionadas[(columnas_seleccionadas["Diabetes_binary"] == 0) & (columnas_seleccionadas['Sex'] == 1)]
no_diabetes_hombres


```

**YOU CAN ANALYZE THE DIFFERENCES BETWEEN CERTAIN PARAMETERS IN EACH DETERMINED GROUP**
